````python
import os
import pandas as pd
from sqlalchemy import create_engine, text
from dotenv import load_dotenv
from google import genai  # Make sure you've installed: pip install google-genai

# âœ… Load environment variables
load_dotenv("../config/credentials.env")

# âœ… Extract environment variables
uri = os.getenv("URI")
token = os.getenv("API")

# âœ… Sanity check (optional, helps debug)
print("Database URI:", uri)
print("Google API Token loaded:", bool(token))

# âœ… Initialize Google GenAI client
client = genai.Client(api_key=token)

# âœ… Connect to PostgreSQL database
engine = create_engine(uri)

# âœ… Optional: Quick test query
with engine.connect() as conn:
    result = conn.execute(text("SELECT version();"))
    print("Connected to:", result.fetchone()[0])

# âœ… Example: load a table into a DataFrame
# df = pd.read_sql("SELECT * FROM your_table_name LIMIT 10;", engine)
# print(df.head())
````

````python
# â”€â”€â”€ 3ï¸âƒ£  Helper: get database schema â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def get_table_schema():
    """Fetch table names + columns to give model context"""
    schema_info = ""
    with engine.connect() as conn:
        tables = conn.execute(text("SELECT table_name FROM information_schema.tables WHERE table_schema='public';")).fetchall()
        for (table_name,) in tables:
            cols = conn.execute(text(f"SELECT column_name FROM information_schema.columns WHERE table_name='{table_name}';")).fetchall()
            col_list = [c[0] for c in cols]
            schema_info += f"\nTable {table_name}: {', '.join(col_list)}"
    return schema_info

# â”€â”€â”€ 4ï¸âƒ£  Generate SQL from question â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def nlp_to_sql(question, schema):
    prompt = f"""
    You are an expert data analyst. Convert the question below into a valid PostgreSQL SQL query.
    Use ONLY tables and columns listed here:
    {schema}

    QUESTION: {question}

    Return ONLY the SQL query (no explanation).
    """

    res = client.models.generate_content(model="gemini-2.0-flash", contents=prompt)
    sql_text = res.text.strip()

    # ðŸ§¹ Clean up common formatting issues
    sql_text = sql_text.replace("```sql", "").replace("```", "").strip()

    # Optional: remove stray backticks or markdown formatting
    sql_text = sql_text.strip("`")

    return sql_text

# â”€â”€â”€ 5ï¸âƒ£  Run SQL safely â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def execute_query(sql_query):
    """Execute the SQL and return a DataFrame or error message."""
    try:
        with engine.connect() as conn:
            df = pd.read_sql(text(sql_query), conn)
        return df
    except Exception as e:
        return f"âŒ SQL Execution Error: {e}"

# â”€â”€â”€ 6ï¸âƒ£  Summarize results â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def summarize_results(question, df):
    """Turn raw SQL output into a readable summary using Gemini."""
    if isinstance(df, str):
        return df  # error message

    # Convert result to markdown table for clarity
    table_md = df.to_markdown(index=False)

    prompt = f"""
    You are a helpful business data assistant.
    The user asked: "{question}"

    Here are the query results:
    {table_md}

    Write a clear, professional summary in simple English that highlights key findings,
    trends, or totals. Avoid SQL language; speak like you're explaining it to a business person.
    """
    res = client.models.generate_content(model="gemini-2.0-flash", contents=prompt)
    return res.text.strip()

# â”€â”€â”€ 7ï¸âƒ£  Main interactive loop â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
if __name__ == "__main__":
    print("ðŸ” Reading PostgreSQL schema...")
    schema = get_table_schema()
    print("âœ… Schema loaded!")

    while True:
        q = input("\nAsk your business question (or 'exit'): ").strip()
        if q.lower() in ["exit", "quit"]:
            break

        sql = nlp_to_sql(q, schema)
        print("\nðŸ§  Generated SQL:\n", sql)

        df = execute_query(sql)
        summary = summarize_results(q, df)

        print("\nðŸ“Š Summary:\n", summary)
````
